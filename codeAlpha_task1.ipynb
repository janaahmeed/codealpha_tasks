SMOTE requires a fully numeric, NaN-free feature matrix.
Regex-based cleaning is fragile; pd.to_numeric(errors="coerce") is the correct and robust approach before imputation and scaling.

```python
#fill naan
def clean_and_impute(series):
    # Convert everything to string, mark anything non-numeric as NaN
    # "coerce" literally means:
    #Handles everything non-numeric

    cleaned = pd.to_numeric(series, errors="coerce")

    # Impute with median
    return cleaned.fillna(cleaned.median())

for col in continous_cols:
    data[col] = clean_and_impute(data[col])

data[continous_cols].isna().sum()
```

```python
from sklearn.preprocessing import RobustScaler
#Scale correctly (all columns together)
scaler = RobustScaler()
data[continous_cols] = scaler.fit_transform(data[continous_cols])
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(
    data[selected_features.index],
    data[target]
)
```

```python

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

estimators = [
   ('gb', GradientBoostingClassifier(n_estimators=50, random_state=42)),
  ('rf', RandomForestClassifier(n_estimators=200, random_state=42))
]

clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(class_weight="balanced"),
    cv=10
)

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split (data[selected_features.index],data[target],test_size=0.2,random_state=42)
```

```python
# 1️⃣ Train the stacking model # 2️⃣ Predict # Evaluate
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
